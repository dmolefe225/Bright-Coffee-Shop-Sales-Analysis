```sql
----------------------------------------------------------------------------------------------------
--RETURNING ALL THE TABLE COLUMNS
----------------------------------------------------------------------------------------------------
SELECT *
FROM `workspace`.`bright_coffee`.`bright_coffee_shop_analysis_case_study_1`
LIMIT 1000;

-----------------------------------------------------------------------------------------------------
--CHECKING FOR THE MAX AND MIN DATE ON OUR DATASET
-----------------------------------------------------------------------------------------------------
SELECT MIN(transaction_date) AS Start_date,
       MAX(transaction_date) AS End_date
FROM `workspace`.`bright_coffee`.`bright_coffee_shop_analysis_case_study_1`
LIMIT 1000;

-------------------------------------------------------------------------------------------------------
--CHECKING FOR DIFFERENT STORE LOCATIONS 
-------------------------------------------------------------------------------------------------------
SELECT DISTINCT store_location
FROM `workspace`.`bright_coffee`.`bright_coffee_shop_analysis_case_study_1`
LIMIT 1000;

------------------------------------------------------------------------------------------------------
--CALCULATING THE TOTAL REVENUE 
-------------------------------------------------------------------------------------------------------
SELECT SUM(transaction_qty*unit_price) AS Total_revenue
FROM `workspace`.`bright_coffee`.`bright_coffee_shop_analysis_case_study_1`
LIMIT 1000;

--------------------------------------------------------------------------------------------------------
--CHECKING WHICH PRODUCT CATEGORY DO WE HAVE ON OUR STORES 
--------------------------------------------------------------------------------------------------------
SELECT DISTINCT product_category
FROM `workspace`.`bright_coffee`.`bright_coffee_shop_analysis_case_study_1`
LIMIT 1000;

---------------------------------------------------------------------------------------------------------
--CHECKING FOR MAX AND MIN SPEND BUCKET 
---------------------------------------------------------------------------------------------------------
SELECT MIN(transaction_qty*unit_price) AS `LOW SPEND`,
       MAX(transaction_qty*unit_price) AS `HIGH SPEND`
FROM `workspace`.`bright_coffee`.`bright_coffee_shop_analysis_case_study_1`;

---------------------------------------------------------------------------------------------------------
--EXCUTING THE LONG QUERY FOR MY ANALYSIS
----------------------------------------------------------------------------------------------------------
SELECT DISTINCT store_location,
       product_category,
       transaction_id,
       SUM(transaction_qty*unit_price) AS total_revenue,
       transaction_date AS deal_date,
       dayname(transaction_date) AS Day_name,
       monthname(transaction_date) AS Month_name,

-- FORMING A NEW COLUMN FOR TIME FRAME

CASE 
           WHEN date_format(transaction_time,'HH:mm:ss') BETWEEN '06:00:00' AND '09:59:59' THEN 'Morning'
           WHEN date_format(transaction_time,'HH:mm:ss') BETWEEN '10:00:00' AND '12:59:59' THEN 'Lunch'
           WHEN date_format(transaction_time,'HH:mm:ss') BETWEEN '13:00:00' AND '15:59:59' THEN 'Afternoon'
           WHEN date_format(transaction_time,'HH:mm:ss') >= '16:00:00' THEN 'Evening'
      END AS time_frame,

-- FORMING A NEW COLUMN FOR DAY CLASSIFICATION
      
 CASE 
      WHEN dayname(transaction_date) IN ('Sun','Sat') THEN 'Weekend'
           ELSE 'Weekday'
      END AS Day_Class,

-- FORMING A NEW COLOMN FOR SPEND BUCKET 

 CASE 
      WHEN (transaction_qty*unit_price) <=50 THEN '01.Low spend'
      WHEN (transaction_qty*unit_price) BETWEEN 51 AND 150 THEN '02.Medium spend'
      WHEN (transaction_qty*unit_price) BETWEEN 150 AND 250 THEN '03.High spend'
      ELSE '04.Expensive'
 END AS Spend_bucket,

-- COUNTING OF IDs

      COUNT(*) AS number_of_sales,
      COUNT(DISTINCT product_id) AS number_of_product,
      COUNT(DISTINCT store_id) AS number_of_stores

FROM `workspace`.`bright_coffee`.`bright_coffee_shop_analysis_case_study_1`
GROUP BY transaction_id,
         transaction_date,
         transaction_time,
         product_id,
         store_id,
         store_location,
         product_category,
         (transaction_qty*unit_price);


----------------------------------------------------------------------------------------------------------------
```
